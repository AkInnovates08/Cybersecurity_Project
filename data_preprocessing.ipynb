```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import joblib

# Step 1: Load dataset
data = pd.read_csv('nslkdd.csv')

# Step 2: Identify categorical columns
categorical_columns = ['protocol_type', 'service', 'flag']  # Example columns

# Step 3: Encode categorical columns
le = LabelEncoder()
for col in categorical_columns:
    data[col] = le.fit_transform(data[col])

# Step 4: Encode the label column
data['label'] = le.fit_transform(data['label'])

# Step 5: Split features and label
X = data.drop('label', axis=1)
y = data['label']

# Step 6: Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 7: Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 8: Initialize and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 9: Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Step 10: Evaluate the model's performance
print("✅ Model Training Completed Successfully!")
print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Step 11: Save the trained model
joblib.dump(rf_model, 'random_forest_model.pkl')
joblib.dump(scaler, 'scaler.pkl')

print("✅ Model Saved Successfully!")
```
